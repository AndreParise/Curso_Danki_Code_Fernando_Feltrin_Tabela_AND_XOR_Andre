````python
entradas = np.array([1, 9, 5])  # Define um array com os valores de entrada (inputs do perceptron)
# Exemplo: Imagine que estamos classificando frutas com base em tr√™s caracter√≠sticas:
# 1 -> Cor (1 significa "vermelha")
# 9 -> Tamanho (9 significa "grande")
# 5 -> Peso (5 significa "m√©dio")

pesos = np.array([0.8, 0.1, 0]) # Define os pesos (import√¢ncia de cada entrada na decis√£o)
# Os pesos determinam o impacto de cada entrada no c√°lculo final:
# 0.8 -> A cor tem um peso alto na decis√£o
# 0.1 -> O tamanho tem um peso pequeno
# 0   -> O peso da fruta n√£o influencia na decis√£o (peso = 0 significa que √© irrelevante)
````

## ** üîé O que est√° acontecendo aqui? **

Criamos um vetor de entradas [1, 9, 5], que simula as informa√ß√µes que o neur√¥nio recebe.

Criamos um vetor de pesos [0.8, 0.1, 0], que define a import√¢ncia de cada entrada no c√°lculo da sa√≠da.

üîπ Exemplo: Imagine que queremos prever se um estudante passar√° na prova com base em tr√™s fatores: quantidade de horas estudadas (1), participa√ß√£o nas aulas (9), e uso de material de apoio (5). Os pesos representam a influ√™ncia de cada fator na aprova√ß√£o.

---
---

````python
def soma(e, p): # Define uma fun√ß√£o chamada 'soma' que recebe dois par√¢metros: 'e' (entradas) e 'p' (pesos)
  return e.dot(p) # Retorna o produto escalar (multiplica√ß√£o elemento a elemento e soma dos resultados)
````

## ** üîé O que est√° acontecendo aqui? **

A fun√ß√£o soma(e, p) recebe as entradas e e os pesos p, multiplicando cada entrada pelo seu peso correspondente e somando os resultados.

e.dot(p) √© o produto escalar, que calcula essa soma de forma eficiente.

üîπ Exemplo:
Se tivermos:

e = [1, 9, 5]
p = [0.8, 0.1, 0]

O c√°lculo ser√°:
(1 * 0.8) + (9 * 0.1) + (5 * 0) = 0.8 + 0.9 + 0 = 1.7

---
---

````python
s = soma(entradas, pesos) # Chama a fun√ß√£o 'soma' passando as vari√°veis 'entradas' e 'pesos' como argumentos e armazena o resultado na vari√°vel 's'
````

Aqui, chamamos a fun√ß√£o soma() e armazenamos o resultado na vari√°vel s. No nosso exemplo, s = 1.7.

---
---

````python
def stepFunction(soma):  # Define uma fun√ß√£o chamada 'stepFunction' que recebe um valor 'soma' como entrada
  if (soma >= 1):  # Verifica se o valor de 'soma' √© maior ou igual a 1
    return 1  # Se for maior ou igual a 1, retorna 1
  return 0  # Caso contr√°rio, retorna 0
````

## ** üîé O que est√° acontecendo aqui? **

A stepFunction() √© a fun√ß√£o de ativa√ß√£o, que define se o neur√¥nio ser√° ativado ou n√£o.

Se soma >= 1, retorna 1 (ativado ‚úÖ).

Caso contr√°rio, retorna 0 (n√£o ativado ‚ùå).

üîπ Exemplo:
Se s = 1.7, o resultado ser√° 1, ativando o neur√¥nio.
Se s = 0.5, o resultado seria 0, n√£o ativando o neur√¥nio.

---
---

````python
saida = stepFunction(s)  # Chama a fun√ß√£o 'stepFunction' passando 's' como argumento e armazena o resultado na vari√°vel 'saida'
````

````python
print(s)     # Exibe o valor da vari√°vel 's' no console (resultado da soma ponderada)
print(saida) # Exibe o valor da vari√°vel 'saida' no console (resultado da fun√ß√£o stepFunction)
````

###  O que √© e para que serve?  
- A **taxa de aprendizado** (`learning rate`) determina **o quanto os pesos v√£o mudar** a cada atualiza√ß√£o durante o treinamento.  
- Um valor **muito alto** pode fazer com que o modelo oscile e n√£o aprenda corretamente.  
- Um valor **muito baixo** pode fazer com que o modelo demore muito para convergir.  

### üìå Exemplo pr√°tico:
Se tivermos um erro na sa√≠da do Perceptron e precisarmos ajustar os pesos, a f√≥rmula de atualiza√ß√£o geralmente √©:  

\[novo¬†peso=peso¬†atual+(taxa¬†de¬†aprendizado√óerro√óentrada)

erro ‚ùå ‚Üí Diferen√ßa entre o valor esperado e o valor calculado pelo Perceptron.
entrada üî¢ ‚Üí O valor da entrada correspondente ao peso que estamos ajustando.
üîÑ Como funciona?
Se o Perceptron errou, o peso ser√° ajustado para se aproximar da resposta correta. A taxa de aprendizado define o "passo" dessa corre√ß√£o.

````python
def funcao_degrau(soma):  # Define uma fun√ß√£o chamada 'funcao_degrau' que recebe um valor 'soma'
  if (soma >= 1):  # Se o valor da soma for maior ou igual a 1
    return 1  # Retorna 1 (ativa√ß√£o do neur√¥nio)
  return 0  # Caso contr√°rio, retorna 0 (neur√¥nio inativo)
````

## üîé**Explica√ß√£o passo a passo:**


### üîç O que esse c√≥digo faz?
1. **Define a fun√ß√£o `funcao_degrau`**, que recebe um n√∫mero chamado `soma` üì•  
2. **Verifica se `soma` √© maior ou igual a 1** üîç  
   - Se for, retorna `1` ‚ö° (o neur√¥nio √© ativado)  
   - Caso contr√°rio, retorna `0` ‚ùå (o neur√¥nio fica inativo)

Se chamarmos a fun√ß√£o com diferentes valores de `soma`:


Isso significa que **apenas valores maiores ou iguais a 1 ativam o neur√¥nio** üöÄ.

---

### üìñ O que √© a fun√ß√£o degrau?
A **fun√ß√£o degrau (Step Function)** √© uma fun√ß√£o de ativa√ß√£o usada em redes neurais.  
Ela **decide se o neur√¥nio ser√° ativado ou n√£o** com base em um limite (neste caso, 1).  

- üü¢ **Se `soma >= 1` ‚Üí O neur√¥nio √© ativado (sa√≠da = 1)**
- üî¥ **Se `soma < 1` ‚Üí O neur√¥nio N√ÉO √© ativado (sa√≠da = 0)**

---

### üéØ Resumo:
- A **fun√ß√£o degrau** √© usada em **Perceptrons** para tomar decis√µes üìä  
- Ela **compara um valor de entrada (`soma`) com um limiar (1)**  
- Se **ultrapassar o limiar**, a sa√≠da ser√° **1** (ativo)  
- Se **n√£o ultrapassar**, a sa√≠da ser√° **0** (inativo)  

üîπ **Exemplo real:**  
Pense na ativa√ß√£o de um interruptor de luz üí°:  
- Se a for√ßa no bot√£o for suficiente üîò ‚Üí A luz acende (1)  
- Se n√£o for suficiente ‚ùå ‚Üí A luz n√£o acende (0)  

Assim, o perceptron toma decis√µes baseadas nessa regra simples! ‚ö°

````python
def calculo_saida(reg):  # Define uma fun√ß√£o chamada 'calculo_saida' que recebe 'reg' como entrada (registro de entrada)
  s = reg.dot(pesos)  # Calcula o produto escalar entre o vetor de entrada 'reg' e o vetor de 'pesos'
  return funcao_degrau(s)  # Passa o resultado 's' para a fun√ß√£o de ativa√ß√£o 'funcao_degrau' e retorna o resultado
````

Essa fun√ß√£o implementa o **aprendizado de um Perceptron de camada √∫nica** usando o **algoritmo de aprendizado supervisionado**.  
Ela ajusta os **pesos do modelo** at√© que o Perceptron classifique corretamente todas as amostras.

## üìå **O que esse c√≥digo faz?**
Ele treina um **Perceptron Simples**, que √© um **modelo de neur√¥nio artificial** usado para resolver problemas de classifica√ß√£o bin√°ria (ex: Porta AND, OR, NOT).  

O processo segue essas **etapas**:
1. **Inicia o erro total** como `1` para come√ßar o treinamento üéØ

3. Para cada entrada do conjunto de treinamento:
   - Calcula a sa√≠da com os pesos atuais üßÆ  
   - Compara com a sa√≠da esperada ‚ùå  
   - **Se houver erro, ajusta os pesos** usando a **Regra Delta** üõ†Ô∏è  
4. **Repete at√© que todas as sa√≠das sejam classificadas corretamente** ‚úÖ

> **Novo Peso = Peso Atual + (Taxa de Aprendizado √ó Entrada √ó Erro)**

````python
entradas = np.array([[0,0],   # Entrada 1
                     [0,1],   # Entrada 2
                     [1,0],   # Entrada 3
                     [1,1]])  # Entrada 4

saidas = np.array([[0],  # Sa√≠da esperada para [0,0] ‚Üí 0
                   [1],  # Sa√≠da esperada para [0,1] ‚Üí 1
                   [1],  # Sa√≠da esperada para [1,0] ‚Üí 1
                   [0]]) # Sa√≠da esperada para [1,1] ‚Üí 0

pesos0 = np.array([[-0.424, -0.740, -0.961],  # Pesos da camada de entrada para a camada oculta
                   [0.358, -0.577, -0.469]])


pesos1 = np.array([[-0.017],  # Pesos da camada oculta para a camada de sa√≠da
                   [-0.893],
                   [0.148]])

````
